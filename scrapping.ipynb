```python
# pip install beautifulsoup4
# pip install requests
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# Extraer datos de un sitio web
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
response = requests.get(url)
response

# Debe dar como respuesta '200' para confirmar la conexi√≥n
```

```python
if response:
    # wb write binary para escribir
    with open('data.csv', 'wb') as dataset:
        dataset.write(response.content)
```

```python
url_1 = "https://en.wikipedia.org/wiki/Gun_death_and_violence_in_the_United_States_by_state#Gun_death_rates"
response_1 = requests.get(url_1)
response_1
```

```python
# Se obtiene el c√≥digo de todo el sitio web en una linea de texto
# b' = es un contenido binary ASCI
response_1.content
```

```python
if response_1:
    soup = BeautifulSoup(response_1.text, 'html')
    soup
```

```python
table = soup.find('table', {'class': 'wikitable'})
table
```

```python
rows = table.find_all('tr')
data = []
for row in rows:
    cells = row.find_all(['td', 'th'])
    cells = [cell.get_text(strip = True) for cell in cells]
    data.append(cells)
```

```python
data
```

```python
df = pd.DataFrame(data)
df

# Casi listo, se debe convertir la primera fila en columnas y despues eliminarla
```

```python
# Extraer toda la fila, las columnas
my_colums = df.loc[0, :].to_list()
my_colums
```

```python
# Renonbrar las columnas
df.columns = my_colums
df
```

```python
# por seguridad se copia la BD
df_clean = df.copy()
# Desde la linea 1 en adelante
df_clean = df_clean[1:]
df_clean
```
