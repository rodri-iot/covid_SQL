```python
# Imports libraries
import pandas as pd
from pandasql import sqldf
```

```python
# Import csv
df = pd.read_csv("https://covid.ourworldindata.org/data/owid-covid-data.csv")
df.info()
```

```python
df.sample(12)
```

```python
# Using SQL Query display the fist 10 rows
sqldf('SELECT *  FROM df LIMIT 10')
```

```python
# Query the first rows of the data for Peru
sqldf("SELECT * FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# Query the first rows of Peru data for the date 2020-07-01
sqldf("SELECT * FROM df WHERE location = 'Peru' AND date = '2020-07-01' LIMIT 10")
```

```python
# Query the first rows of Peru data for the date 2020-07-01 to 2020-07-10
sqldf("SELECT * FROM df WHERE location = 'Peru' AND date >= '2020-07-01' AND date <= '2020-07-10' LIMIT 10")
```

```python
# Query the first 10 rows of Peru data for the date 2020-07-01 to 2020-07-10, only select columns: location, date, total_cases, total_deaths
sqldf("SELECT location, date, total_cases, total_deaths FROM df WHERE location = 'Peru' AND date >= '2020-07-01' AND date <= '2020-07-10' LIMIT 10")
```

```python
# Query for display unique location
sqldf("SELECT DISTINCT location FROM df LIMIT 10")
```

```python
# Query for display unique locations order by desc
sqldf("SELECT DISTINCT location FROM df ORDER BY location DESC LIMIT 30")
```

```python
# Query 
sqldf("SELECT SUM(new_deaths) FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# QUERY
sqldf("SELECT MAX(population) FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# Query 
sqldf("SELECT SUM(new_deaths) FROM df WHERE location = 'World' LIMIT 10")
```

```python
# Query 
sqldf("SELECT SUM(new_deaths), MAX(population) FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# Query 
sqldf("SELECT SUM(new_deaths) AS DEATHS, MAX(population) AS POPULATION FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# Query 
sqldf("SELECT SUM(new_deaths)/MAX(population) AS Ratio FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# Query 
sqldf("SELECT SUM(new_deaths)/MAX(population) AS Ratio FROM df WHERE location = 'World' LIMIT 10")
```

```python
# Query 
sqldf("SELECT (SUM(new_deaths)/MAX(population))*100 AS Ratio FROM df WHERE location = 'Peru' LIMIT 10")
```

```python
# Query 
sqldf("SELECT (SUM(new_deaths)/MAX(population))*100 AS Ratio FROM df WHERE location = 'World' LIMIT 10")
```
